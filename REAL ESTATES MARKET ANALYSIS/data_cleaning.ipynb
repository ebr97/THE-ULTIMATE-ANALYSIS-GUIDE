### Creating the data frame that contain all the data from all .csv files

```python
path = './CSV'
files = [file for file in os.listdir(path) if not file.startswith('.')] # ignore hidden files

df = pd.DataFrame()

for file in files:
    data = pd.read_csv(path + '/' + file)
    df = pd.concat([df, data])
```

From the details column we will make more columns that store individual data

Add built_area column

```python
def get_built_area(built_area):
    return built_area.split(' ')[0][2:].strip() # we will split after space because there are some areas that have decimal number marked with ,

df['built_area'] = df['details'].apply(lambda x: get_built_area(x))
df
```

Add usable_area column

```python
def get_usable_area(usable_area):
    return usable_area.split(' ')[2].strip()  # we will split after space because there are some areas that have decimal number marked with ,

df['usable_area'] = df['details'].apply(lambda x: get_usable_area(x))

df
```

Add type column to see if the flats are new or not.

```python
def get_type(type):
    #return type.split('2')[1].split(',')[1].strip() # we will split after 2 because is present in each any line (because our dataset contains only 2 rooms flats) and it refears to the number of rooms
    for item in type.split(','):
        if item in [" 'locuinta noua'"," 'nou'"]:
            return 'nou'
        elif item in [" 'utilizat'" ," 'locuinta utilizata'", " 'renovat'"]:
            return 'utilizat'

df['type'] = df['details'].apply(lambda x: get_type(x))


df
```

Let's add the year column

We will split after the '2', if the lenght of list resulted is 3  we will take the first value then will be adding the 2 upfront, if not we will replace the value with default one '>2000'

```python
def get_year(year):
    #if len(year.split('2')) == 3:
        #return '2' + year.split('2')[2][0:3]
    
    for item in year.split():
        if item in ["'2004',", "'2005',", "'2006',", "'2007',", "'2008',", "'2009',", "'2010',", "'2011',", "'2012',", "'2013',", "'2014',", "'2015',", "'2016',", "'2017',", "'2018',", "'2019',", "'2020',", "'2021',", "'2022',", "'2023',"]:
            item = item.rstrip(',')
            item = item.rstrip("'")
            item = item.lstrip("'")
            return item
       

df['year'] = df['details'].apply(lambda x: get_year(x))

df
```

Add the floor column

```python
def get_floor(floor):
    return floor.split()[-1]

df['floor'] = df['details'].apply(lambda x: get_floor(x))
df
```

Let's add the orientation

```python
def get_orientation(orientation):
    if len(orientation.split('2')) == 3:
        if len(orientation.split('2')[2].split(',')) > 1:
            return orientation.split('2')[2].split(',')[1] # case year is 2000 - 2019 (without 2002)
    elif len(orientation.split('2')) == 4:
        if len(orientation.split('2')[3].split(',')) > 1:
            return orientation.split('2')[3].split(',')[1] # case year is  2020 or 2021
    elif len(orientation.split('2')) == 5:
        if len(orientation.split('2')[4].split(',')) > 1:
            return orientation.split('2')[4].split(',')[1] # case year is 2022
    else:
        return 'other'

df['orientation'] = df['details'].apply(lambda x: get_orientation(x))
df
```

As we can see some columns display data incorrectly so we need to adjust that.<br>
Let's escape strings from aditional charcters (''])

```python
# creating functions
def remove_left_ampersand(string):
    return string.lstrip("'")

def remove_right_ampersand(string):
    return string.rstrip("'")

def remove_right_comma(string):
    return string.rstrip(",")

def remove_right_parenthesis(string):
    return string.rstrip("]")
```

Cleaning the usable_area column

```python
df['usable_area'] = df['usable_area'].apply(lambda x: remove_left_ampersand(x))
df['usable_area'] = df['usable_area'].apply(lambda x: remove_right_comma(x))
df['usable_area'] = df['usable_area'].apply(lambda x: remove_right_ampersand(x))


df
```

Cleaning type column

```python
df['type'] = df['type'].apply(lambda x: x.strip())
df['type'] = df['type'].apply(lambda x: remove_left_ampersand(x))
df['type'] = df['type'].apply(lambda x: remove_right_ampersand(x))

df
```

Cleaning  floor column

```python
df['floor'] = df['floor'].apply(lambda x: remove_left_ampersand(x))
df['floor'] = df['floor'].apply(lambda x: remove_right_parenthesis(x))
df['floor'] = df['floor'].apply(lambda x: remove_right_ampersand(x))

df
```
